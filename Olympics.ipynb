## Basically the previously value we got that is wrong 
## there gold of a NOC means sum of all golds won by each player but there are some games where a team of players played a game and own a single medal but our data is adding all medals so the total number of medals own by a NOC becomes wrong
## So we thought an approach if we drop the duplicated rows then we can't find medal own by a player so we thought to add three extra columns where we can store the medals own by team not by each players


```python
# Identify the first occurrence of a team's win
# (Returns True for the 1st player, False for all other teammates)
is_original_win = ~summer_olympics.duplicated(subset=['NOC', 'Year', 'Season', 'Event', 'Medal'])
is_original_win
# Create the "Official" columns by multiplying by the mask
summer_olympics['team_bronze'] = summer_olympics['Bronze'] * is_original_win
summer_olympics['team_gold'] = summer_olympics['Gold'] * is_original_win
summer_olympics['team_silver'] = summer_olympics['Silver'] * is_original_win

```

### Based on NOC Medal counts

```python
df = summer_olympics.groupby(['NOC','region']).sum()[["team_gold", "team_silver", "team_bronze"]].sort_values("team_gold", ascending = False)
```

### We can find details of match for a 'particular year' and for a 'particular region'

```python
medal_df = summer_olympics.drop_duplicates(['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
def medal_tally(Year,region):
    flag = 0
    if(Year == "Overall" and region == "Overall"):
        temp_df = medal_df
    if(Year == "Overall" and region != "Overall"):
        flag = 1
        temp_df = medal_df[medal_df['region'] == region]
    if(Year != "Overall" and region == "Overall"):
        temp_df = medal_df[medal_df['Year'] == int(Year)]
    if(Year != "Overall" and region != "Overall"):
        temp_df = medal_df[(medal_df['Year'] == int(Year)) & (medal_df['region'] == region)]

    if(flag == 1):
        print(temp_df.groupby('Year').sum()[['team_bronze','team_gold','team_silver','total_team_medal']].sort_values('total_team_medal'))
    else:
        print(temp_df.groupby(['NOC','region']).sum()[['team_bronze','team_gold','team_silver','total_team_medal']].sort_values('total_team_medal'))
    
```

```python
print(medal_tally('Overall','India'))
```

```python
print(medal_tally(2016,'Overall'))
```
