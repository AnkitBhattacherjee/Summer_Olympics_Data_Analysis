## Team medal alocation

```python
# Define your list of team sports
Team_sports = [
    'Football', 'Basketball', 'Hockey', 'Volleyball', 
    'Beach Volleyball', 'Water Polo', 'Handball', 
    'Rugby', 'Rugby Sevens', 'Baseball', 'Softball', 
    'Tug-Of-War', 'Polo'
]

# team_game
df_team_events = df_copy[df_copy['Sport'].isin(Team_sports)]

# non-team_game
df_non_team_events = df_copy[~df_copy['Sport'].isin(Team_sports)]


is_original_win = ~df_team_events.duplicated(subset = ['Event','Year','Team','NOC','Medal'])

# team_game gold, bronze, silver count
df_team_events['Team Bronze'] = df_team_events['Bronze']*is_original_win
df_team_events['Team Gold'] = df_team_events['Gold']*is_original_win
df_team_events['Team Silver'] = df_team_events['Silver']*is_original_win
df_team_events['Team Gold'] = df_team_events['Team Gold'].astype(int)
df_team_events['Team Silver'] = df_team_events['Team Silver'].astype(int)
df_team_events['Team Bronze'] = df_team_events['Team Bronze'].astype(int)

# non_team_game gold, bronze, silver count
df_non_team_events['Team Bronze'] = df_non_team_events['Bronze']
df_non_team_events['Team Gold'] = df_non_team_events['Gold']
df_non_team_events['Team Silver'] = df_non_team_events['Silver']

# df_team_events
# df_non_team_events

# final dataframe after concatination of the two tables
final_df = pd.concat([df_team_events, df_non_team_events])
final_df
```
